## ***Implementation:***

```
%tensorflow_version 2.x
import tensorflow as tf
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

```
number_instances = [0] * 10
images = np.zeros([10, 28, 28], dtype = np.uint32)

# Counting how many times each number appears and summing up all the train-images.
for i in range(len(train_images)):
  number_instances[train_labels[i]] += 1
  images[train_labels[i]] += train_images[i]

# Loop that find the average of each number.
for i in range(10):
  images[i] //= number_instances[i]
```

```
# Displaying the output
for i in range(10):
  plt.figure(figsize=(4, 4))
  plt.imshow(images[i])
  plt.title(f"Number of apperances: {number_instances[i]}")
  plt.axis('off')
  plt.show()
```
