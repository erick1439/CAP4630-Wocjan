#  **Problem 1:**

## ***Description:***

Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that

p1 is a minimum of f1

p2 is a maximum of f2

p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector.

Use (a) three-dimensional plots as in three_dimensional_plotting.ipynb and (b) density plots and (c) contour plots as in density_and_contour_plots.ipynb. Indicate clearly the points p1, p2, and p3 in all plots.

## ***Implementation:***

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
#plt.style.use('seaborn-white')
```

#### Bivariate function with minimun point. 
#### $f1 = x^{2} + y^{2},\ p1 = (0,0)$

```
def f1(x,y): 
  return (x**2 + y**2)

x = np.linspace(-4,4,30)
y = np.linspace(-4,4,30)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(50)
ax.contour3D(X, Y, Z, 200)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlabel('z')
ax.scatter3D(0, 0, 0, cmap='Greens')
```

```
plt.contour(X, Y, Z, 30)
plt.plot(0, 0, 'o', c='green')
plt.colorbar()
plt.show()
```

```
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.plot(0, 0, 'o', c='green')
plt.colorbar()
plt.show()
```

#### Bivariate function with maximun point.

#### $f2 = cos(x + y) + 1,\ p2 = (0,0)$

```
def f2(x, y):
  return np.cos(x + y) + 1

x = np.linspace(-3.5, 3.5, 30)
y = np.linspace(-3.5, 3.5, 30)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(5, 120)
ax.contour3D(X, Y, Z, 100)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_zlabel('z')
ax.scatter3D(0, 0, 2, cmap='Greens')
```

```
plt.contour(X, Y, Z, 40)
plt.plot(0, 0, 'o', c='green')
plt.colorbar()
plt.show()
```

```
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.plot(0, 0, 'o', c='green')
plt.show()
```

#### Bivariate function with gradint as the zero vector.
#### $f3 = cos(x) + sin(y),\ p3 = (0, - \frac{\pi}{2})$

```
def f3(x, y):
  return np.cos(x) + np.sin(y)

x = np.linspace(-4, 4, 30)
y = np.linspace(-4, 4, 30)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(45, 45)
ax.contour3D(X, Y, Z, 100)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter3D(0, -np.pi / 2, cmap="Greens" )
```

```
plt.contour(X, Y, Z, 30)
plt.plot(0, -np.pi / 2, 'o', c='green')
plt.colorbar()
plt.show()
```

```
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.plot(0, -np.pi / 2, 'o', c='green')
plt.show()
```
